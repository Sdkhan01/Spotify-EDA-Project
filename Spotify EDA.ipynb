## Data Analysis

### Top 5 most popular artists

```python
df.columns
```

```python
top_five_artist = df.groupby("artist").count().sort_values(by="song_title", ascending= False )["song_title"][:5]
top_five_artist
```

**Now Plotting the above data**

```python
top_five_artist.plot.barh() #barh() will plot the bars horizontally
plt.show()
```

### Results:

* **Most popular artist is Drake, followed by Rick Ross, Disclosure, Backstreet Boys, Walk The Moon.**

### Top 5 loudest tracks

```python
loudest_tracks = df[['loudness', 'song_title']].sort_values(by = 'loudness',ascending = True )[:5] #Here we took asceding = True because loudness will increase with negative decibals
loudest_tracks
```

```python
plt.figure(figsize=(12, 7)) #Setting the size of the figure
sns.barplot(x="loudness", y="song_title", data= loudest_tracks)
plt.title("Top 5 loudes tracks")
plt.show()
```

### Results : 

* **Here we can see,  Lyric Pieces Arietta, is loudest song,  followed by #8 Fantasiestcke Des Abends,  #Piano Quartet in E flat Andante cantabile,  #I Was So Young and You Were So Beautiful and  #Eleanor.***

### Top 5 Energetic Tracks 

```python
energetic_tracks = df[['energy', 'song_title']].sort_values(by = 'energy',ascending = False )[:5] 
energetic_tracks
```

```python
plt.figure(figsize=(12, 5)) #Setting the size of the figure
sns.barplot(x="energy", y="song_title", data= energetic_tracks)
plt.title("Top 5 energetic tracks")
plt.show()
```


### Results :
* **Most Energetic track is No Absolution followed by  #Pitchblack,  #A Darkened Gospel, #Unanswered, #First Contact.**

### Artist with the most danceability song

```python
danceable_songs = df[['danceability', 'song_title', 'artist']].sort_values(by = 'danceability' , ascending = False )[:5]
danceable_songs
```

**Now Visualizing the data using plt,sns.**



```python
plt.figure(figsize= (12,5))
sns.barplot(x = 'danceability', y = 'artist', data = danceable_songs)
plt.title("Artists With The Most Danceable Song ")
plt.show()
```

 ### Results : 
 **Most danceable artists is Ben Remember, followed by Justin Timberlake, Blaqstarr, Young Thug, Vanilla Ice**

### To 5 instrumentalness tracks

```python
top_instrumentals = df[['instrumentalness', 'song_title', 'artist']].sort_values(by= 'instrumentalness', ascending= False)[:5]
top_instrumentals
```

**Now Visualizing the data using plt.pie :**

```python
plt.figure(figsize=(12,8))
plt.pie(x = 'instrumentalness', data = top_instrumentals, autopct= '%1.2f%%', labels=top_instrumentals.song_title)[:10]
plt.show()

##autopct : This format allow us to discplay the percent value using string formatting
```

### Result :
* **Senseless order is most leading instrumental track with 20.24% , followed by So Destroyed (channeling Rage Peace)(20.07%)	, A Hundred Dreams from Now(19.99%), Raven (19.86%), Strandbar(disko) 19.84%**

### Multiple Feature Plots:

```python
feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature in feature_cols:
    pos_data = df[df['target']== 1][feature] #Positive data
    neg_data = df[df['target']== 0][feature] #Negative data 
    
    plt.figure(figsize= (12,7))
    
    sns.histplot(pos_data, bins= 30, label = 'Positive', color= 'green')
    sns.histplot(neg_data, bins= 30, label = 'Negative', color= 'maroon')
    
    plt.legend(loc= 'upper right')
    plt.title(f'Positive and Negative Historgram Plot For {feature}')
    plt.show()
```
